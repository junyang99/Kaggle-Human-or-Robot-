```python
import pandas as pd

bids_df = pd.read_csv('data/bids.csv')
train_df = pd.read_csv('data/train.csv')
test_df = pd.read_csv('data/test.csv')
```

```python
bidder_unique = bids_df.groupby("bidder_id").nunique()
bidder_unique = bidder_unique.drop(columns='bidder_id', errors='ignore').reset_index()
```

```python
train_set = train_df.merge(bidder_unique, on='bidder_id', how='left')
test_set = test_df.merge(bidder_unique, on='bidder_id', how='left')
```

```python
train_set = train_set.fillna(0)
test_set = test_set.fillna(0)
```

```python
train_set
```

```python
test_set
```

```python
bids_work = train_df.merge(bids_df, on='bidder_id',how='left')
bids_work
```

```python
bids_work_test = test_df.merge(bids_df, on='bidder_id',how='left')
bids_work_test
```

```python
test = bids_work.groupby('bidder_id').nunique()
test
```

```python
bots_data = bids_work[bids_work.outcome == 1]
bots_data.head()
```

```python
bots_data = bots_data.fillna('idk')
bots_data
```

```python
country_df = bots_data
total_bots = country_df['outcome'].sum()
country_unique = country_df.groupby('country').sum() / total_bots
country_unique = country_unique.drop(columns=['bid_id','time'], errors='ignore')
country_unique.columns=['bots_per_country']
country_unique.head()
```

```python
new_df = bids_work.merge(country_unique, how='left', on='country')
new_df.fillna(0)
new_df.head()
```

```python
new_test_df = bids_work_test.merge(country_unique, how='left', on='country')
new_test_df.fillna(0)
new_test_df.head()
```

```python
final_df = new_df.groupby('bidder_id').mean()
final_df = final_df.drop(columns=['outcome','bid_id','time'])
final_df.columns=['mean_bots_per_country']
final_df
```

```python
test_final_df = new_test_df.groupby('bidder_id').mean()
test_final_df = test_final_df.drop(columns=['bid_id','time'],errors='ignore')
test_final_df.columns=['mean_bots_per_country']
test_final_df
```

```python
train_set = train_set.merge(final_df, on='bidder_id')
train_set
```

```python
test_set = test_set.merge(test_final_df, on='bidder_id', how='left')
test_set
```

```python
#feature 'response_time'
bids_response_time = bids_df.sort_values(by=['time'])
bids_response_time['response_time'] = bids_response_time.groupby('auction')['time'].diff()
mean = bids_response_time.groupby("bidder_id")['response_time'].mean()
train_set = train_set.merge(mean, on='bidder_id', how='left')
test_set = test_set.merge(mean, on = 'bidder_id', how='left')
# train_set
```

```python
#feature 'bids_per_auction'
train_set['auction'] = train_set['auction'].replace(to_replace=0.0,value=1.0)

def bids_per_auction(datarow):
    return (datarow['bid_id'] / datarow['auction'])

train_set['bids_per_auction'] = train_set.apply(bids_per_auction, axis=1)
# train_set

test_set['auction'] = test_set['auction'].replace(to_replace=0.0,value=1.0)

def bids_per_auction(datarow):
    return (datarow['bid_id'] / datarow['auction'])

test_set['bids_per_auction'] = test_set.apply(bids_per_auction, axis=1)
```

```python
#feature ip_per_bid
train_set['ip_per_bid'] = train_set['ip'] / train_set['bid_id']
test_set['ip_per_bid'] = test_set['ip'] / test_set['bid_id']
train_set['ip_per_bid'].fillna(train_set['ip_per_bid'].mean())
```

```python
#feature device_per_bid
train_set['device_per_bid'] = train_set['device'] / train_set['bid_id']
test_set['device_per_bid'] = test_set['device'] / test_set['bid_id']
train_set['device_per_bid'].fillna(train_set['device_per_bid'].mean())
```

```python
#feature auction_per_device
train_set['auction_per_device'] = train_set['auction'] / train_set['device']
test_set['auction_per_device'] = test_set['auction'] / test_set['device']
train_set['auction_per_device'].fillna(train_set['auction_per_device'].mean())
```

```python
#feature auction_per_device
train_set['response_time_per_device'] = train_set['response_time'] / train_set['device_per_bid']
test_set['response_time_per_device'] = test_set['response_time'] / test_set['device_per_bid']
train_set['response_time_per_device'].fillna(train_set['response_time_per_device'].mean())
```

```python
#feature IP_per_device
train_set['IP_per_device'] = train_set['ip'] / train_set['device']
test_set['IP_per_device'] = test_set['ip'] / test_set['device']
train_set['IP_per_device'].fillna(train_set['IP_per_device'].mean())
```

```python
#feature Auction_per_devicebid
#tried and decrease AUC
train_set['device_per_auctionbid'] = train_set['device_per_bid'] / train_set['auction_per_device']
test_set['device_per_auctionbid'] = test_set['device_per_bid'] / test_set['auction_per_device']
train_set['device_per_auctionbid'].fillna(train_set['device_per_auctionbid'].mean())
```

```python
# #feature Auction_per_devicebid
# #tried and decrease AUC
# train_set['merchandise_per_response_time'] = train_set['merchandise'] / train_set['response_time']
# test_set['merchandise_per_response_timed'] = test_set['merchandise'] / test_set['response_time']
# train_set['merchandise_per_response_time'].fillna(train_set['merchandise_per_response_time'].mean())
```

```python
#Additional url than auction
#accuracy no change but decreased AUC by 0.02
train_set['additional_url'] = train_set['url'] - train_set['auction']
test_set['additional_url'] = test_set['url'] - test_set['auction']
train_set['additional_url'].fillna(0)
```

```python
#Additional bid per time period
#improved accuracy by 0.67 but decreased AUC by 0.015
train_set['additional_bid'] = train_set['bid_id'] - train_set['time']
test_set['additional_bid'] = test_set['bid_id'] - test_set['time']
train_set['additional_bid'].fillna(0)
```

```python
#Additional bid per time period
#improved accuracy by 0.67 but decreased AUC by 0.015
train_set['bid_id_per_country'] = train_set['bid_id'] - train_set['country']
test_set['bid_id_per_country'] = test_set['bid_id'] - test_set['country']
train_set['bid_id_per_country'].fillna(0)
```

```python
##Additional feature for auction_per_bid_id ???
train_set['sum_auction_id'] = train_set['auction'] * train_set['bid_id']
test_set['sum_auction_id'] = test_set['auction'] * test_set['bid_id']
train_set['sum_auction_id'].fillna(0)
```

```python
test_set = test_set.fillna(0)
train_set = train_set.fillna(0)
```

```python
import plotly.express as px
corr = train_set.corr()
fig = px.imshow(corr)
fig.update_xaxes(side="top")
fig.show()
```

```python
# features = ['mean_bots_per_country',
# 'response_time',
# 'ip_per_bid',
# 'response_time_per_device',
# 'IP_per_device',
# 'device_per_auctionbid',
# 'additional_url',
# 'additional_bid',
# 'sum_auction_id',
# 'bid_id',
# 'auction',
# 'merchandise',
# 'device',
# 'time',
# 'country',
# 'ip',
# 'url']

features = ['bids_per_auction','device_per_bid','bid_id','device_per_auctionbid','ip_per_bid',
            'mean_bots_per_country','response_time']

X_kaggle = test_set[features]
```

```python
#TTS Testing for GB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

X = train_set[features]
y = train_set['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=5)

model = GradientBoostingClassifier(learning_rate=0.01,random_state=5)
model.fit(X_train, y_train)

y_predict = model.predict(X_test)
y_predict_proba = model.predict_proba(X_test)
y_predict_proba_output = y_predict_proba[:, 1]

accuracy_test = metrics.accuracy_score(y_test, y_predict)
print('\nAccuracy_score on test dataset : ', accuracy_test)

accuracy = metrics.accuracy_score(y_test, y_predict)
auc = metrics.roc_auc_score(y_test, y_predict_proba_output)

print('Accuracy:', round(accuracy, 4) * 100, '%')
print('AUC:', round(auc, 4))
```
